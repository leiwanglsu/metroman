```python
import pandas as pd

# Load the USGS RDB file
file_path = r'C:\Users\leiwang\Documents\ArcGIS\Projects\SWOT_project\data\SWOT_Alabama\usgs_rdb.txt'  # Replace with the actual file path
data = pd.read_csv(file_path, sep='\t', comment='#', header=1)

# Ensure the date column is correctly parsed
# Adjust 'datetime_column' and 'value_column' to match the actual column names in your file
data['datetime'] = pd.to_datetime(data['20d'])
data['date'] = data['datetime'].dt.date  # Extract only the date part

# Calculate the daily mean values
daily_mean = data.groupby('date')['14n'].mean() *  0.028316846592

# Save or print results
daily_mean = pd.DataFrame(daily_mean)
daily_mean.to_csv(r"C:\Users\leiwang\Documents\ArcGIS\Projects\SWOT_project\data\SWOT_Alabama\daily_mean_values.csv", index=True)
```
